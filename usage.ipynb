```python
import json
from PIL import Image, ImageDraw
```

```python
def overlay_bboxes(pil_img, bboxes):
    draw = ImageDraw.Draw(pil_img)
    for (xmin, ymin, xmax, ymax) in bboxes:    
        # draw bounding box
        shape = [(xmin, ymin), (xmax, ymax)] 
        draw.rectangle(shape, fill=None, outline=(255,255,255), width=2)
        
    return pil_img
```

```python
with open("annotations_xyxy.json") as f:
    data_xyxy = json.loads(f.read())
```

```python
fnames = data_xyxy['file_name']
annotations_xyxy = data_xyxy['bboxes_xyxy']
```

```python
img = Image.open('images/'+fnames[0])
overlay_bboxes(img, annotations_xyxy[0])
```

```python
img = Image.open('images/'+fnames[10])
overlay_bboxes(img, annotations_xyxy[10])
```

```python
img = Image.open('images/'+fnames[23])
overlay_bboxes(img, annotations_xyxy[23])
```
